```python
def generateParenthesis(n: int) -> list[str]:
    dp = [[] for _ in range(n + 1)]
    dp[0] = [""]

    for k in range(1, n + 1):
        for i in range(k):
            for inner in dp[i]:
                for outer in dp[k - 1 - i]:
                    dp[k].append(f"({inner}){outer}")

    return dp[n]
```
