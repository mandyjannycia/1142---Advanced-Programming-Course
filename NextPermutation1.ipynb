```python
def nextPermutation(nums: list[int]) -> None:
    n = len(nums)

    # Step 1: find the rightmost pivot where nums[i] < nums[i+1]
    pivot = -1
    for i in range(n - 2, -1, -1):
        if nums[i] < nums[i + 1]:
            pivot = i
            break

    # Step 2: if no pivot, array is fully descending → just reverse
    if pivot == -1:
        nums.reverse()
        return

    # Step 3: find rightmost element greater than pivot and swap
    for j in range(n - 1, pivot, -1):
        if nums[j] > nums[pivot]:
            nums[pivot], nums[j] = nums[j], nums[pivot]
            break

    # Step 4: reverse the suffix after pivot to get smallest order
    left, right = pivot + 1, n - 1
    while left < right:
        nums[left], nums[right] = nums[right], nums[left]
        left += 1
        right -= 1
```
