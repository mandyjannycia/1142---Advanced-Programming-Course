```python
def removeElement(nums: list[int], val: int) -> int:
    left, right = 0, len(nums) - 1

    while left <= right:
        if nums[left] == val:
            nums[left] = nums[right]  # overwrite with last element
            right -= 1
        else:
            left += 1

    return left
```
