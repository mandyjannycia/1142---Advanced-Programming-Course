```python
def reverse(x: int) -> int:
    INT_MIN, INT_MAX = -(2**31), 2**31 - 1
    sign = -1 if x < 0 else 1
    x = abs(x)

    result = 0
    while x != 0:
        digit = x % 10
        x //= 10
        result = result * 10 + digit

    result *= sign
    return result if INT_MIN <= result <= INT_MAX else 0
```
