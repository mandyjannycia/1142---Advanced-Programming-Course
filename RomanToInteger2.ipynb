```python
def romanToInt(s: str) -> int:
    replacements = [
        ("IV", "IIII"), ("IX", "IIIII"),
        ("XL", "XXXX"), ("XC", "XXXXX"),
        ("CD", "CCCC"), ("CM", "CCCCC")
    ]
    for pair, replacement in replacements:
        s = s.replace(pair, replacement)

    roman = {'I': 1, 'V': 5, 'X': 10, 'L': 50,
             'C': 100, 'D': 500, 'M': 1000}

    return sum(roman[c] for c in s)
```
