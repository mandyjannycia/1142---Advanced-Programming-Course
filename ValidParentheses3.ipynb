```python
def isValid_optimized(s: str) -> bool:
    if len(s) % 2 != 0:
        return False

    bracket_pairs = {'(': ')', '{': '}', '[': ']'}
    stack = []

    for char in s:
        if char in bracket_pairs:
            stack.append(char)
        else:
            if not stack:
                return False
            if bracket_pairs[stack.pop()] != char:
                return False

    return len(stack) == 0
```
